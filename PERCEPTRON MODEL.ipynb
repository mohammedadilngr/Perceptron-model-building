```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from sklearn.linear_model import Perceptron
```

```python
data=pd.read_excel('CHURNDATA.xlsx')
```

```python
data=data[['AGE','CUS_Month_Income','total transactions','Status']]
```

```python
sns.scatterplot(x=data['AGE'],y=data['total transactions'],hue=data['Status'])
```

```python
data.fillna(0,inplace=True)
```

```python
def Perceptron(AGE,CUS_Month_Income,total_transactions):
    Status=12088*AGE-68800.14*CUS_Month_Income+315751*total_transactions+246
    if Status>=0:
        Status='Active'
    else:
        Status='Churn'
        
    return Status
```
