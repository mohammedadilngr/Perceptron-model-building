```python
def Perceptron(X,y):
    X=X.values
    y=y.values
    X=np.insert(X,0,1,axis=1)
    w0,w1,w2,w3=1,1,1,1
    n=0.001
    for i in range(0,len(X)):
        ind=np.random.randint(0,len(X))
        y_pred=w0*X[ind][0]+(w1*X[ind][1])+(w2*X[ind][2])+(w3*X[ind][3])
        def step(y_pred):
            if y_pred >=0:
                y_pred=1
            else :
                y_pred=0
            return y_pred
        y_pred=step(y_pred)
        w0=w0-n*(y[ind]-y_pred)*X[ind][0]
        w1=w1-n*(y[ind]-y_pred)*X[ind][1]
        w2=w2-n*(y[ind]-y_pred)*X[ind][2]
        w3=w3-n*(y[ind]-y_pred)*X[ind][3]
        weight=np.array([w0,w1,w2,w3])
    return weight
```
